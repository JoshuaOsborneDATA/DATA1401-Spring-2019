** Exercise 3: ** For the top 3 continuous observables, draw a Reciever Operating Characteristic (ROC) curve and compute the area under the curve. 

```python
########## continous means between 2 values, so using,thalach,trestbps,chol############
#!pip install scikit-learn --user
import copy
from sklearn.metrics import roc_curve, auc
df_heart_copy=copy.deepcopy(df_heart)
df_heart_copy.loc[df_heart_copy.cp != 0, 'cp'] = 1
#print df_heart_copy

fpr, tpr,_= roc_curve(df_heart_copy.cp,df_heart_copy.thalach )
roc_auc = auc(fpr, tpr)
plt.plot(fpr,tpr,color='red',label='ROC curve (area = %0.2f)' % roc_auc)
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('fps vs. tpr for thalach')
plt.show()

fpr, tpr,_= roc_curve(df_heart_copy.cp,df_heart_copy.trestbps )
roc_auc = auc(fpr, tpr)
plt.plot(fpr,tpr,color='red',label='ROC curve (area = %0.2f)' % roc_auc)
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('fps vs. tpr for trestbps')
plt.show()

fpr, tpr,_= roc_curve(df_heart_copy.cp,df_heart_copy.chol )
roc_auc = auc(fpr, tpr)
plt.plot(fpr,tpr,color='red',label='ROC curve (area = %0.2f)' % roc_auc)
plt.legend()
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('fps vs. tpr for chol')
plt.show()
```
